```python
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_excel('Dataset_Q4.xlsx')
#seed_centroids = data.iloc[:4]
data.head()
```

```python
plt.scatter(data['X'], data['Y'])
```

```python
kmeans2 = KMeans(n_clusters=2)
y_kmeans2 = kmeans2.fit_predict(data)

newcentroids = kmeans2.cluster_centers_

# visualizing the formed clusters
plt.scatter(data['X'], data['Y'],c=y_kmeans2, cmap="rainbow") 
plt.scatter(newcentroids[:, 0], newcentroids[:, 1], c='black')
```

```python
# performing z score normalization
from scipy.stats import zscore
zdata = data.apply(zscore)
zdata.head()
```

```python
z_kmeans2 = KMeans(n_clusters=2)
z_y_kmeans2 = z_kmeans2.fit_predict(zdata)

newcentroids = z_kmeans2.cluster_centers_

# visualizing the formed clusters
plt.scatter(zdata['X'], zdata['Y'],c=z_y_kmeans2, cmap="rainbow") 
plt.scatter(newcentroids[:, 0], newcentroids[:, 1], c='black')
```

```python
# performing min max scaling only

from sklearn import preprocessing

x = data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
mmdata = pd.DataFrame(x_scaled)
mmdata.head()
```

```python
mm_kmeans2 = KMeans(n_clusters=2)
mm_y_kmeans2 = mm_kmeans2.fit_predict(mmdata)

newcentroids = mm_kmeans2.cluster_centers_

# visualizing the formed clusters
plt.scatter(mmdata[0], mmdata[1],c=mm_y_kmeans2, cmap="rainbow") 
plt.scatter(newcentroids[:, 0], newcentroids[:, 1], c='black')
```

```python
# performing min max scaling to z score normalized dataset

x = zdata.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
mmdata = pd.DataFrame(x_scaled)
print(mmdata.head())

mm_kmeans2 = KMeans(n_clusters=2)
mm_y_kmeans2 = mm_kmeans2.fit_predict(mmdata)

newcentroids = mm_kmeans2.cluster_centers_

# visualizing the formed clusters
plt.scatter(mmdata[0], mmdata[1],c=mm_y_kmeans2, cmap="rainbow") 
plt.scatter(newcentroids[:, 0], newcentroids[:, 1], c='black')
```
